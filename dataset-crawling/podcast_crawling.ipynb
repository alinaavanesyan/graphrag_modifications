```python
from pydub import AudioSegment
import speech_recognition as sr


audio = AudioSegment.from_file('audio.wav')
chunks = [audio[i:i + 30000] for i in range(0, len(audio), 30000)]

recognizer = sr.Recognizer()
final_text = ''

for idx, chunk in enumerate(chunks):
    chunk.export(f'chunk_{idx}.wav', format='wav')
    with sr.AudioFile(f'chunk_{idx}.wav') as source:
        audio_data = recognizer.record(source)
    try:
        text = recognizer.recognize_google(audio_data, language="ru-RU")
        final_text+=' '
        final_text+=text
        print(f"Транскрипт для куска {idx}:", text)
    except sr.UnknownValueError:
        print(f"Не удалось распознать речь для куска {idx}")
    except sr.RequestError as e:
        print(f"Ошибка распознавания речи для куска {idx}: {e}")
```
